# 01 — Dataset Validation

Experimentation only — loads each dataset through `brandparadigm.datasets` and inspects schema/label distribution. See `docs/dataset_guide.md` for provisioning instructions (Amazon Reviews / Reddit are local files; TweetEval downloads live).

```python
from brandparadigm.config.paths import CONFIGS_DIR
from brandparadigm.datasets import load_dataset, NoLocalDataError
from brandparadigm.utils import read_yaml

config = read_yaml(CONFIGS_DIR / "data_config.yaml")
```

```python
tweeteval_df = load_dataset("tweeteval", config, split="test", sample_size=500)
print(tweeteval_df.shape)
tweeteval_df["label"].value_counts()
```

```python
try:
    amazon_df = load_dataset("amazon", config)
    print(amazon_df.shape)
    display(amazon_df.head())
except NoLocalDataError as exc:
    print(exc)
```

```python
try:
    reddit_df = load_dataset("reddit", config)
    print(reddit_df.shape)
    display(reddit_df.head())
except NoLocalDataError as exc:
    print(exc)
```
